```python
#Importing necessary Libraries for Data Manipulation & Visualization

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading datasets for analysis of Tracks

df_tracks = pd.read_csv('features.csv', encoding='cp1252')
df_tracks.head()
```

```python
#Checking Null values in Dataset

pd.isnull(df_tracks).sum()
```

```python
#Viewing no. of rows & columns and also checking whether they are null or not

df_tracks.info()
```

```python
#Sorting: Least Popular Songs on Spotify

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#Applying descriptive statistics on dataset

df_tracks.describe().transpose()
```

```python
#Finding: Top 10 most popular songs on Spotify

most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#Finding: A particular artist in a row

df_tracks[["artist_name"]].iloc[18]
```

```python
#Converting: ms into seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
#Visualization: Correlation Map

corr_df = df_tracks.drop(["key","mode"], axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))

heatmap = sns.heatmap(corr_df, annot= True, fmt= ".1g", vmin= -1, vmax= 1, center= 0, cmap= "inferno", linewidths= 1, linecolor= "Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation= 90)
```

```python
#Sampling out 0.4% of original dataframe

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#Plotting: Regression plot between Loudness & Energy

plt.figure(figsize= (10,6))
sns.regplot(data= sample_df, y= "loudness", x= "energy",color="c").set(title= "Loudness vs Energy Correlation")
```

```python
#Plotting: Regression plot between Popularity & Acousticness

plt.figure(figsize= (10,6))
sns.regplot(data= sample_df, y= "popularity", x= "acousticness",color="b").set(title= "Popularity vs Acousticness Correlation")
```

```python
#Loading datasets for Genre analysis

df_genre = pd.read_csv('features.csv', encoding='cp1252')
df_tracks.head()
```

```python
#Plotting: Bar Plot between Genre & Duration

plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y="genre", x="duration_ms", data= df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
# Plotting: Bar Plot between Popularity & Genres (out of Top 10 in Dataset)

sns.set_style(style= "darkgrid")
plt.figure(figsize= (10,5))
famous = df_genre.sort_values("popularity", ascending= False).head(10)
sns.barplot(y="genre", x="popularity", data= famous).set(title="Top 5 Genres by Popularity")
```
